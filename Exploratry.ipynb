# Data analysis
**Import library and adding data in Panda**

```
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np


'''downlaod haberman.csv from https://www.kaggle.com/gilsousa/habermans-survival-data-set/version/'''
#Load haberman.csv into a pandas dataFrame.
haberman = pd.read_csv("haberman.csv")
print("Shape", haberman.shape)
print("All columns", haberman.columns)
haberman.describe()
```

**Observation:**
*   Dataset contains shape 306 Row & 4 column
*   4 Column are age,year,nodes,status
*  Patient age gorup is 30 to 83 years
*   Year between 1958 to 1969
*   Positive nodes between 0 to 52
*   Survival Status 1 = Patient suvive 5 years or more, 2 = Patient died within 5 years










# **Scatter Plot**
**1-D scatter plot**

```
import numpy as np
sns.set_style("whitegrid")
haberman_survival = haberman.loc[haberman["status"] == 1];
haberman_died = haberman.loc[haberman["status"] == 2];
plt.plot(haberman_survival["age"], np.zeros_like(haberman_survival['age']), 'o')
plt.plot(haberman_died["age"], np.zeros_like(haberman_died['age']), 'o')
plt.xlabel('Age')
plt.legend(['Survival More than 5 Years','Survival Less than 5 Years'], shadow=True, fancybox=True,facecolor='C8')
plt.title('1-D Scatter Plot- Age',loc='center',pad='20',fontsize='20',color='Blue',backgroundcolor='Orange')
plt.show()
```

**Observation:**
Age group 30-33 years patient survive 5 years or more

```
plt.plot(haberman_survival["nodes"], np.zeros_like(haberman_survival['nodes']), 'o')
plt.plot(haberman_died["nodes"], np.zeros_like(haberman_died['nodes']), 'o')
plt.legend('12')
plt.xlabel('Nodes')
plt.title('1-D Scatter Plot- Positive Nodes',loc='center',pad='20',fontsize='20',color='Blue',backgroundcolor='Orange')
plt.legend(['Survival More than 5 Years','Survival Less than 5 Years'], shadow=True, fancybox=True,facecolor='C8')
plt.show()
```

**Observation:**
Very dense points

```
plt.plot(haberman_survival["year"], np.zeros_like(haberman_survival['year']), 'o')
plt.plot(haberman_died["year"], np.zeros_like(haberman_died['year']), 'o')
plt.legend('12')
plt.xlabel('year')
plt.title('1-D Scatter Plot- Year',loc='center',pad='20',fontsize='20',color='Blue',backgroundcolor='Orange')
plt.legend(['Survival More than 5 Years','Survival Less than 5 Years'], shadow=True, fancybox=True,facecolor='C8')
plt.show()
```

**Observation:** 
Each year patient found who died in 5 years

# **Univariate Analysis**

**Historam,PDF & CDF**

```
sns.FacetGrid(haberman, hue="status", size=5) \
   .map(sns.distplot, "age") \
   .add_legend();
plt.title('Historam with Age',loc='center',pad='20',fontsize='20',color='Blue',backgroundcolor='Orange')
plt.legend(['>5 Years','<5 Years'], shadow=True)
plt.show();
```

**Observation:**
Patient Age between 30 to 34 survive more than 5 years

```
sns.FacetGrid(haberman, hue="status", size=5) \
   .map(sns.distplot, "nodes") \
   .add_legend();
plt.title('Historam with Nodes',loc='center',pad='20',fontsize='20',color='Blue',backgroundcolor='Orange')
plt.legend(['>5 Years','<5 Years'], shadow=True)
plt.show();
```

**Observation:**
Patient who has less than 2 positive auxilary nodes found there is chances to survive more than 5 years

**PDF & CDF**

```
counts, bin_edges = np.histogram(haberman_survival['age'], bins=10, 
                                 density = True)
pdf = counts/(sum(counts))
cdf = np.cumsum(pdf)
plt.xlabel('age')
plt.title('PDF & CDF of AGE in the Group of Survival patient',loc='center',pad='20',fontsize='20',color='Blue',backgroundcolor='Orange')
plt.plot(bin_edges[1:],pdf);
plt.plot(bin_edges[1:], cdf)
```

**Observation:**
80% patient survive more than 5 years whose age is less than 62 years



```
counts, bin_edges = np.histogram(haberman_survival['year'], bins=10, 
                                 density = True)
pdf = counts/(sum(counts))
#print(pdf);
#print(bin_edges);
cdf = np.cumsum(pdf)
plt.title('PDF & CDF of Year in the Group of Survival patient',loc='center',pad='20',fontsize='20',color='Blue',backgroundcolor='Orange')
plt.plot(bin_edges[1:],pdf);
plt.plot(bin_edges[1:], cdf)
plt.xlabel('year')
```

```
counts, bin_edges = np.histogram(haberman_survival['nodes'], bins=10, 
                                 density = True)
pdf = counts/(sum(counts))
cdf = np.cumsum(pdf)
plt.title('PDF & CDF of Nodes in the Group of Survival patient',loc='center',pad='20',fontsize='20',color='Blue',backgroundcolor='Orange')
plt.plot(bin_edges[1:],pdf);
plt.plot(bin_edges[1:], cdf)
plt.xlabel('nodes')
```

**Observation:**
All most 81% patient who has less than 4 positive nodes are survive more than 5 years

```
counts, bin_edges = np.histogram(haberman_died['year'], bins=10, 
                                 density = True)
pdf = counts/(sum(counts))
cdf = np.cumsum(pdf)
plt.title('PDF & CDF of YEAR in the Group of Died patient within 5 years',loc='center',pad='20',fontsize='20',color='Blue',backgroundcolor='Orange')
plt.plot(bin_edges[1:],pdf);
plt.plot(bin_edges[1:], cdf)
plt.xlabel('year')
```

**Observation:**
80% patient died before 1966

```
counts, bin_edges = np.histogram(haberman_died['age'], bins=10, 
                                 density = True)
pdf = counts/(sum(counts))
cdf = np.cumsum(pdf)
plt.title('PDF & CDF of Age in the Group of Died patient within 5 years',loc='center',pad='20',fontsize='20',color='Blue',backgroundcolor='Orange')
plt.plot(bin_edges[1:],pdf);
plt.plot(bin_edges[1:], cdf)
plt.xlabel('age')
```

**Observation:**
80% patient died between age of 34 to 62

```
counts, bin_edges = np.histogram(haberman_died['nodes'], bins=10, 
                                 density = True)
pdf = counts/(sum(counts))
cdf = np.cumsum(pdf)
plt.title('PDF & CDF of Nodes in the Group of Died patient within 5 years',loc='center',pad='20',fontsize='20',color='Blue',backgroundcolor='Orange')
plt.plot(bin_edges[1:],pdf);
plt.plot(bin_edges[1:], cdf)
plt.xlabel('nodes')
```

**Observation:**
Positivi nodes less than 10 survival chance are more than 5 years
